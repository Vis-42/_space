```python
import cmath
```

```python
def heron(a,b,c):
    s=(a+b+c)/2
    return cmath.sqrt(s*(s-a)*(s-b)*(s-c))
```

```python
a,b,c=3,7,12
print(f'the area of the triangle is {heron(a,b,c):.5f}')
```
